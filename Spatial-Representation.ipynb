```python
import matplotlib.pyplot as plt
import matplotlib.patches as pt
```

```python
fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(1,1,1)
circle1 = pt.Circle( (15, 15), 1, facecolor="green", edgecolor="#000000FF",
                   linewidth = 1.0, linestyle = "-", alpha=0.5, zorder=2)
circle2 = pt.Circle( (15, 15), 5**0.5, facecolor="green", edgecolor="#000000FF",
                   linewidth = 1.0, linestyle = "-", alpha=0.5, zorder=1)
circle3 = pt.Circle( (15, 15), 50**0.5, facecolor="green", edgecolor="#000000FF",
                   linewidth = 1.0, linestyle = "-", alpha=0.5, zorder=0)
ax.add_patch(circle1)
ax.add_patch(circle2)
ax.add_patch(circle3)
ax.set_xlim(0, 30)
ax.set_ylim(0, 30)
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, aspect='equal') 
triangle1 = pt.Polygon([[2,2],[ ((2**0.5)/2) + 2 , 2 + (2 ** 0.5) ],[2 + (2 ** 0.5),2]], facecolor="#FFD700", alpha=0.5, edgecolor="black", closed=True,fill=True)
triangle2 = pt.Polygon([[4,2],[((10 ** 0.5)/2) + 4 , 2 + (10 ** 0.5)],[4 + (10**0.5),2]], facecolor="#FFD700", alpha=0.5, edgecolor="black", closed=True,fill=True)
triangle3 = pt.Polygon([[7.75,2],[((100 ** 0.5)/2) + 7.75, 2 + (100 ** 0.5)],[7.75 + (100**0.5),2]], facecolor="#FFD700", alpha=0.5, edgecolor="black", closed=True,fill=True)


ax.add_patch(triangle3)
ax.add_patch(triangle2)
ax.add_patch(triangle1)
ax.set_xlim((0, 20)) 
ax.set_ylim((0, 15)) 
plt.show()
```
